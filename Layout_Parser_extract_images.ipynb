```python
def get_data(images):
  all_figures=[]
  for im in images :
     layout = model.detect(im)
     figure_blocks = lp.Layout([b for b in layout if ((b.type=='Figure'))])
     for block in figure_blocks:
         figure = (block.pad(left=5, right=5, top=5, bottom=5).crop_image(im))
         all_figures.append(figure)
  return all_figures
```
